```python
# spliitting of experience in a understandable way
experience_split = df['experience'].str[0:-1].str.split('-', expand=True)
experience_split.head()
```

```python
# removing yrs to make it meanaingful and removing spaces by strip
experience_split[1] =  experience_split[1].str.strip()
#print(experience_split[1])
experience_split[1] = experience_split[1].str.replace(r'yr', '')
experience_split[1].head()
```

```python
experience_split[0] = pd.to_numeric(experience_split[0], errors='coerce')
experience_split[1] = pd.to_numeric(experience_split[1], errors='coerce')
```

```python
# creating a datadrame of experience
experience=pd.concat([experience_split[0], experience_split[1]], axis=1, sort=False)
```

```python
# renaming o and 1 to min and max experience
experience.rename(columns={0:'min_experience', 1:'max_experience'}, inplace=True)
experience.head()
```

```python
# concatenation of df and experience data frame
df=pd.concat([df, experience], axis=1, sort=False)
df.head()
```

```python
pay=df.payrate
```

```python
pay_split = df['payrate'].str[0:-1].str.split('-', expand=True)
pay_split.sample(10)
```

```python
pay_split[0] =pay_split[0].str.strip()
#print(pay_split[0])
pay_split[0] = pay_split[0].str.replace(',', '')# removes ,
pay_split[0] = pay_split[0].str.replace(r'\D.*', '') # removes other than numbers
pay_split[0].head(5)
```

```python
pay_split[1] = pay_split[1].str.strip()
#print(pay_split[0])
pay_split[1] = pay_split[1].str.replace(',', '')# removes ,
pay_split[1] = pay_split[1].str.replace(r'\D.*', '') # removes other than numbers
pay_split[1].head(5)
```

```python
pay_split[0] = pd.to_numeric(pay_split[0], errors='coerce')
pay_split[1] = pd.to_numeric(pay_split[1], errors='coerce')
```

```python
pay=pd.concat([pay_split[0], pay_split[1]], axis=1, sort=False)
pay.head(10)
```

```python
pay.rename(columns={0:'min_pay', 1:'max_pay'}, inplace=True)
pay.head()
```

```python
df=pd.concat([df,pay],axis=1,sort=False)
df.sample(5)
```

```python
# randomfunction to develop all comparative bar graphs
def plots(x):
    for i in range(len(x)):
        #print(x)
        sns.set(style="whitegrid")
        top=df[x[i]].value_counts().head(15)  
        plt.title('graph of'+' '+ x[i]+' '+'by total count')
        top.plot(kind='bar')
        plt.show()
    return  plt.show()
```
